**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
import matplotlib
'''matplotlib.use('TkAgg') #For interactive plotting'''
matplotlib.use('Agg') #For non-interactive plotting
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
ax1.plot(x,y)
ax2.plot(x,z)

ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_xlim(20,22)
ax1.set_ylim(30,50)
ax1.xaxis.set_major_locator(MultipleLocator(0.5))
ax1.yaxis.set_major_locator(MultipleLocator(5))

ax2.set_xlabel('X')
ax2.set_ylabel('Z')
ax2.set_xlim(0,100)
ax2.set_ylim(0,10000)
ax2.xaxis.set_major_locator(MultipleLocator(20))
ax2.yaxis.set_major_locator(MultipleLocator(2000))

ax1.set_title('zoom')

plt.show()
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.use('Agg') #For interactive plotting
from matplotlib.ticker import MultipleLocator
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
plt.subplot(1,2,1)
plt.plot(x,y, linestyle = 'dashed')

plt.subplot(1,2,2)
plt.plot(x,z, color = 'r')

plt.subplot(1,2,1).set_xlim(0,100)
plt.subplot(1,2,1).set_ylim(0,200)
plt.subplot(1,2,2).set_xlim(0,100)
plt.subplot(1,2,2).set_ylim(0,10000)
plt.subplot(1,2,1).xaxis.set_major_locator(MultipleLocator(25))
plt.subplot(1,2,1).yaxis.set_major_locator(MultipleLocator(25))
plt.subplot(1,2,2).xaxis.set_major_locator(MultipleLocator(25))
plt.subplot(1,2,2).yaxis.set_major_locator(MultipleLocator(2000))

plt.show()
```

**See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.use('Agg') #For interactive plotting
from matplotlib.ticker import MultipleLocator
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
plt.subplots(1,2,figsize=(15,3))
plt.subplot(1,2,1,)
plt.plot(x,y,lw ='5' )

plt.subplot(1,2,2,)
plt.plot(x,z,ls = '--' ,color = 'r',lw= '5')

plt.subplot(1,2,1).set_xlim(0,100)
plt.subplot(1,2,1).set_ylim(0,200)
plt.subplot(1,2,2).set_xlim(0,100)
plt.subplot(1,2,2).set_ylim(0,10000)
plt.subplot(1,2,1).xaxis.set_major_locator(MultipleLocator(25))
plt.subplot(1,2,1).yaxis.set_major_locator(MultipleLocator(25))
plt.subplot(1,2,2).xaxis.set_major_locator(MultipleLocator(25))
plt.subplot(1,2,2).yaxis.set_major_locator(MultipleLocator(2000))

plt.tight_layout()
plt.show()
```
